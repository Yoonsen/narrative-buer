```python
corpus = dh.Corpus(doctype = "digibok", author= "derrida", limit = 10)
corpus.corpus[['urn', 'title','authors','year', 'subjects']]
```

Plot i to runder, en med dh.Dispersion, og en med narration_code som nc, nc.display() - noen ganger henger det skikkelig - andre ganger ligger dh.Dispersion() ca. 10 ganger så lang tid.


```python
%%time
for row in corpus.corpus.iterrows():
    urn = row[1]["urn"]
    title = row[1]['title']
    dh.Dispersion(urn=urn, wordbag=['.'], window=1000, pr=100).plot(title=title, figsize=(12,5))
```
